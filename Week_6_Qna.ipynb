# 4. On average which region has the highest

## a. GDP per capita (current US$)

```python
dt1 = data[['Region','GDP per capita (current US$)']]

gp1 = dt1.groupby('Region')['GDP per capita (current US$)'].mean()

print(gp1)
```

```python
print(gp1.idxmax(),'has the highest GDP per capita (current US$) of',gp1.max())
```

## b. International trade: Imports (million US$)

```python
dt2 = data[['Region','International trade: Imports (million US$)']]

gp2 = dt2.groupby('Region')['International trade: Imports (million US$)'].mean()

print(gp2)
```

```python
print(gp2.idxmax(),'has the highest International trade: Imports (million US$) of',gp2.max())
```

# 5. Find out which region has more consistent with respect to GDP percapita (current US$)

```python
dt3 = data[['Region','GDP per capita (current US$)']]

gp3 = dt3.groupby('Region')['GDP per capita (current US$)'].std()

print(gp3)
```

```python
print(gp3.idxmin(),'has more consistent with respect to GDP percapita (current US$) of',gp3.min())
```
